# 인공신경망 구현하기



# 코드에 주어진 패키지 이외에는 추가로 패키지를 사용하지 않습니다.


```
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split
iris = load_iris()
```

# 학습용 데이터로 인공신경망을 학습시키고 테스트용 데이터로 성능을 검증
feature값은 꽃잎 길이, 꽃잎 너비, 꽃받침 길이, 꽃받침 너비입니다.

ans값은 꽃의 품종입니다. (0: setosa, 1:versicolor )


ㅁ 학습용 데이터:

X_train,
y_train

ㅁ 테스트용 데이터:

X_test,
y_test



```
feature = iris.data[:100]
ans = iris.target[:100]

# training set& test set 생성
X_train, X_test, y_train, y_test = train_test_split(feature, ans, test_size=0.3, random_state=1)     # training set sample: 70개 / test set sample: 30개 from feature & ans
y_train = y_train.reshape(70, 1)
print(y_train.shape)
# X_train (70,4) X_test (30,4) y_train (70,) y_test (30,)

# 가중치 설정
w = np.random.uniform(0, 1, (4, 1))
b = np.random.uniform(0, 1, (1, 1))
# print(w, b)

learning_rate = 0.0001
epochs = 40                    # 최적화

# tau(x) = x   선형함수 사용
pred = X_train.dot(w) + b
print(pred.shape)

dw = ((pred - y_train)*X_train).sum(0)
db = (pred - y_train).sum(0)
dw = dw.reshape(4,1)
print(dw.shape)

# 가중치 갱신
w -= learning_rate * dw
b -= learning_rate * db
```

## 학습방향만 맞다면 인공신경망 학습에 필요한 사항들은 자유롭게 구현하시면 됩니다.
(ex. 목적함수, 하이퍼파라미터, 최적화방법 등)

```
for epoch in range(epochs):
####################   인공신경망을 학습시키는 코드를 작성하세요   ############################
  pred = X_train.dot(w) + b
  dw = ((pred - y_train)*X_train).sum(0)
  db = (pred - y_train).sum(0)
  dw = dw.reshape(4,1)

  w -= learning_rate * dw
  b -= learning_rate * db

  if epoch % (epochs // 10) == 0:
    err = np.sum(np.square(pred - y_train))
    print("error: ", err)
    print("w: ", w.tolist())
    print("b: ", b, "\n")

    final_w = w
    final_b = b 

print("final_w:", final_w.tolist())
print("final_b:", final_b)
# print("final_w:", w.tolist())
# print("final_b:", b)
###############################################################################################
```

## 성능을 검증할 때, 정확도를 출력하도록 하세요 (정확도: 30개의 test sample중 예측이 맞은 개수) 

```
#################   인공신경망 모형의 성능을 검증하는 코드를 작성하세요   #####################

prediction = X_test.dot(final_w) + final_b            # X_test와 최종 가중치를 내적
print(prediction.tolist())

# y_test와 값 비교를 위해 반올림 처리
prediction = prediction.round()    

# print(type(prediction[0][0]))
# print(type(y_test[0]))

# y_test와 prediction 인덱스들의 타입 통일하기
print(y_test.astype(np.float64))                    
# print(type(y_test[0]))

# error 값
err = np.sum(np.square(prediction - y_test))
# print(err)

accuracy = 0
for a, b in zip(prediction, y_test):
  if a[0] == b :
    accuracy += 1
    # print(accuracy) 
###############################################################################################


print("accuracy:",accuracy)
```
