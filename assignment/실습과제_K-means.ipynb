# k-Means 알고리즘 구현하기



# 코드에 주어진 패키지 이외에는 추가로 패키지를 사용하지 않습니다.


```
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
iris = load_iris()
import matplotlib.pyplot as plt
import numpy as np
# print(iris)                                   # 'target_names': array(['setosa', 'versicolor', 'virginica']  ==> 예상 클러스터 n = 3
print(iris.data)
```

# 군집분석 결과를 시각화 하기 위해 2차원 데이터로 변형
feature값은 꽃받침 길이, 꽃받침 너비, 꽃잎 길이, 꽃잎 너비 입니다.

X는 feature값에서 꽃잎 길이, 꽃잎 너비 정보만 가져온 데이터입니다.



```
feature = iris.data                              # iris 객체에서 data 객체만 feature에 대입
X = feature[:,[2,3]]                             # feature[2], feature[3]만 빼서 하나의 array로 묶어 X에 대입 
# print(X)
plt.plot(X[:,0], X[:,1], '^')                    # 꽃잎 길이를 x, 꽃잎 너비를 y로 설정, 그래프 그리기
# plt.scatter(X[:,0], X[:,1], marker = 'o')
```

```
model = KMeans(n_clusters=3, n_init=3, max_iter=5, random_state=10).fit(X)

# 센터 정보
c0, c1, c2 = model.cluster_centers_
print(c0)
print(c1)
print(c2)

# 샘플별 클러스터링(군집) 정보
model.labels_
```

```
# plt.scatter(X[:,0],X[:,1], marker = '.')                                                            # 전체 샘플      
plt.scatter(X[model.labels_ == 0,0], X[model.labels_ == 0,1], marker = 'v', facecolor = 'r')          # 0을 배정받은 샘플들 좌표(x, y)
plt.scatter(X[model.labels_ == 1,0], X[model.labels_ == 1,1], marker = '*', facecolor = 'g')          # 1을 배정받은 샘플들 좌표(x, y)
plt.scatter(X[model.labels_ == 2,0], X[model.labels_ == 2,1], marker = 'o', facecolor = 'b')          # 2를 배정받은 샘플들 좌표(x, y)

# centroid 점, (x,y) 좌표 찍기
plt.scatter(c0[0], c0[1], marker = 'v', s = 250, facecolor = 'pink')
plt.scatter(c1[0], c1[1], marker = '*', s = 250, facecolor = 'pink')
plt.scatter(c2[0], c2[1], marker = 'o', s = 250, facecolor = 'pink')
```

## k-Means 알고리즘을 구현하고 이를 시각화하시면 됩니다.
단, 적절한 군집개수를 직접 찾아보세요. 수치적인 결과를 얻을 필요없이 눈으로 확인하는 정도로 군집분석을 수행하면 됩니다.

```
def plot_kmeans(n):
    ###############   코드를 작성하여 ipynb 파일을 제출하세요   ##################
  
  # array(['setosa', 'versicolor', 'virginica'] 3개의 범주를 고려해 n=3으로 
  model = KMeans(n_clusters = 3, init='k-means++', n_init = 5, max_iter = n, random_state = 10).fit(X)
  c0, c1, c2 = model.cluster_centers_

  plt.scatter(X[model.labels_ == 0,0], X[model.labels_ == 0,1], marker = 'v', facecolor = 'r')          
  plt.scatter(X[model.labels_ == 1,0], X[model.labels_ == 1,1], marker = '*', facecolor = 'g')          
  plt.scatter(X[model.labels_ == 2,0], X[model.labels_ == 2,1], marker = 'o', facecolor = 'b')

  plt.scatter(c0[0], c0[1], marker = 'v', s = 250, facecolor = 'pink')
  plt.scatter(c1[0], c1[1], marker = '*', s = 250, facecolor = 'pink')
  plt.scatter(c2[0], c2[1], marker = 'o', s = 250, facecolor = 'pink')

  plt.pause(1)

    ##############################################################################

for i in range(1,6):
    plot_kmeans(i)
```

```
def plot_kmeans(n):
    ###############   코드를 작성하여 ipynb 파일을 제출하세요   ##################
  
  # n이 더 커지면 어떻게 클러스터링이 되는지 궁금해 더 크게 잡아보았습니다.
  model = KMeans(n_clusters = 5, init='k-means++', n_init = 5, max_iter = n, random_state = 10).fit(X)
  c0, c1, c2, c3, c4 = model.cluster_centers_

  plt.scatter(X[model.labels_ == 0,0], X[model.labels_ == 0,1], marker = 'v', facecolor = 'r')          
  plt.scatter(X[model.labels_ == 1,0], X[model.labels_ == 1,1], marker = '*', facecolor = 'g')          
  plt.scatter(X[model.labels_ == 2,0], X[model.labels_ == 2,1], marker = 'o', facecolor = 'b')
  plt.scatter(X[model.labels_ == 3,0], X[model.labels_ == 3,1], marker = 'x', facecolor = 'gray')
  plt.scatter(X[model.labels_ == 4,0], X[model.labels_ == 4,1], marker = '^', facecolor = 'purple')

  plt.scatter(c0[0], c0[1], marker = 'v', s = 250, facecolor = 'pink')
  plt.scatter(c1[0], c1[1], marker = '*', s = 250, facecolor = 'pink')
  plt.scatter(c2[0], c2[1], marker = 'o', s = 250, facecolor = 'pink')
  plt.scatter(c2[0], c2[1], marker = 'x', s = 250, facecolor = 'pink')
  plt.scatter(c2[0], c2[1], marker = '^', s = 250, facecolor = 'pink')

  plt.pause(1)
  

    ##############################################################################

for i in range(1,6):
    plot_kmeans(i)


# 군집을 더 크게 잡아도 결국 3개의 군집(['setosa', 'versicolor', 'virginica'])으로 맞춰진다는 걸 알 수 있었습니다. 
```
