```python
import numpy as np
```

```python
train_x = np.random.uniform(-5, 5, (100, 5))    # 훈련 feature로 -5에서 5까지 랜덤하게 정규분포의 숫자를 뽑아내는데 '100x5'의 행렬 형태로 ==> 가중치는 5개의 값이 있는 행렬 형태
# print(trainx_x)

ans_weight = [1, 2, 3, 4, 4]                    # 가중치 답지, type = list
ans_w = np.array(ans_weight).reshape(5, 1)      # list -> array(배열)
ans_b = 3                                       # bias node 답지
# print(ans_w.shape)

train_y = train_x.dot(ans_w)+ans_b     # 답지 / train_x와 ans_w를 내적(feature와 weight를 각각 곱한 후 합하는 것)
print(train_y.shape)                       # 100개의 값 출력(100,1)



# ans_w를 찾기 위한 과정 시작
w = np.random.uniform(-5, 5, (5, 1))   # 5x1 열벡터로 초기화
b = np.random.uniform(-5, 5, (1, 1))
# print(w, b)

learning_rate = 0.0001        # 학습률
epochs = 200                  # 반복횟수

'''
# tau(x) = x  활성함수로 선형함수 사용
pred = train_x.dot(w) + b    # 모형에서의 출력값 / 그래서 tau(x)의 x로 train_x.dot(w) + b가 들어감
print(pred.shape)

dw = ((pred - train_y)*train_x).sum(0)    # gradient of weight 수식 이해하기 
db = ((pred - train_y)).sum(0)            # gradinet of bias weight
# print(dw.shape)  => (5, )
dw = dw.reshape(5, 1)

# 가중치 갱신
w -= (learning_rate * dw)
b -= (learning_rate * db)
'''

# epochs for loop

for epoch in range(epochs):
  pred = train_x.dot(w) + b
  dw = ((pred - train_y)*train_x).sum(0)
  db = ((pred - train_y)).sum(0)
  dw = dw.reshape(5, 1)

  w -= (learning_rate * dw)
  b -= (learning_rate * db)

  if epoch % (epochs // 20) == 0:     # epoch가 10단위의 숫자일 경우
      err = np.sum(np.square(pred - train_y))
      print("error: ", err)
      print("w: ", w.tolist())
      print("b: ", b, "\n")

print("ans_w: ", ans_w.tolist())
print("final_w: ", w.tolist())
print("ans_b: ", ans_b)
print("final_b: ", b)
```
